```python
import cv2
import numpy as np
import time
import  os
import  threading
import time
```

```python
cap = cv2.VideoCapture(0)
#cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)  # Windows 用户可以试试 DirectShow
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1920)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 1080)
```

```python
#初始化帧计数器和起始时间
frame_count = 0
# start_time的值是时间戳
start_time = time.time()
while True:
    #读取一帧
    ret, frame = cap.read ()
    if not ret:
        print("无法读取视频帧")
        break
    #增加帧计数
    frame_count +=  1

    #计算经过的时间
    elapsed_time = time.time() - start_time
    fps = frame_count / elapsed_time
    # 从左上角截取800*600的区域
    cropped_frame = frame[0:600, 0:800]

    #生成roi区域 该区域用ocr来判断我们进的是哪张地图
    roi_map = cropped_frame[0: 120,600:790]
    #print(f"地图名称{img_name},")
    cv2.imshow("ror",roi_map)
    cv2.rectangle(cropped_frame, (662,0),(783,25),(0,0,255), 1)
    # 为了让观察效果更直观，将当前 FPS 和相机参数信息绘制到图像上
    cv2.putText(cropped_frame, "FPS: {:.2f}".format(fps), (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    print("123")
    #显示图像
    cv2.imshow("yolo", cropped_frame)
    #按q建退出循环
    if cv2.waitKey(30) & 0xff == ord('q'):
        break
#释放视像头资源
cap.release()
#释放opencv
cv2.destroyAllWindows()
```
